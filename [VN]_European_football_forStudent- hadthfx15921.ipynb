## Câu hỏi 6: Liệt kê các trận đấu và thông tin về leage và country tương ứng

Bảng Match có quan hệ với :
 - Bảng Country thông qua khóa country_id
 - Bảng League thông qua khóa League_id
 
Sử dụng câu lệnh join để kết nối dư liệu từ 3 bảng trên.


```python
#write your query
result_country = pd.read_sql("SELECT Match.stage, Match.date, Match.season, League.name as League_name, Country.name as Country_name FROM Match INNER JOIN Country ON Match.country_id=Country.id INNER JOIN League ON Match.country_id=League.id", conn)
result_country
```

## Câu hỏi 7: Liệt kê số lượng trận đấu của mỗi giải đấu bao gồm cả tên giải đấu, sắp xếp theo thứ tự giảm dần số trận đấu

```python
#write your query
result_sum_match = pd.read_sql("Select League.name, sum(Match.stage) as 'sum_stage' from Match inner join League on Match.country_id=League.id group by League.name order by sum_stage desc", conn)
result_sum_match
```

## Câu hỏi 12: Liệt kê tên các đội bóng và tổng số bàn thắng mỗi đội, sắp xếp  theo số lượng giảm dần

- Tổng số bàn thắng của một đội bao gồm cả số bàn thắng ở sân nhà và sân khách.
- Sử dụng câu lệnh Union  

```python
#write your query
result_team_total_goal = pd.read_sql("select (home_team_goal+away_team_goal) as sum_total_goal, Team.team_long_name from Match inner join Team on Team.team_api_id=Match.away_team_api_id union select (home_team_goal+away_team_goal) as sum_total_goal, Team.team_long_name from Match inner join Team on Team.team_api_id=Match.home_team_api_id group by team_long_name order by sum_total_goal desc", conn)
result_team_total_goal
```

## Câu hỏi 14: Liệt kê số trận thắng, thua và hòa của mỗi đội.

- Nếu một đội có số lượng bàn thắng trên sân nhà > số lượng bàn thắng trên sân khách của chính đội đó, thì đội đó sẽ "thắng" trong trận đấu này
- Nếu một đội có số lượng bàn thắng trên sân nhà < số lượng bàn thắng trên sân khách của chính đội đó, thì đội đó sẽ "thua" trong trận đấu này
- Nếu một đội có số lượng bàn thắng trên sân nhà = số lượng bàn thắng trên sân khách của chính đội đó, thì đội đó sẽ "hòa" trong trận đấu này
- Sử dụng câu lệnh "Case When"


```python
#write your query
result_team = pd.read_sql("select sum(away_team_goal) as 'sum_away_team_goal', sum(home_team_goal) as 'sum_home_team_goal', Team.team_long_name from Match inner join Team on Team.team_api_id=Match.away_team_api_id union select sum(away_team_goal) as 'sum_away_team_goal', sum(home_team_goal) as 'sum_home_team_goal', Team.team_long_name from Match inner join Team on Team.team_api_id=Match.home_team_api_id group by team_long_name", conn)
result_team
```

```python
result_team_win=pd.read_sql(
"""select *, 
case 
    when sum_home_team_goal>sum_away_team_goal then 'win'
    when sum_home_team_goal<sum_away_team_goal then 'loss'
    when sum_home_team_goal=sum_away_team_goal then 'draw'
end as 'result'
from result_team""", conn)
result_team_win
```
